```python
import pandas as pd
import yfinance as yf
```

```python
sp500 = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500_df = pd.read_html(sp500)
```

```python
sp500_df
```

```python
len(sp500_df)
```

```python
sp_df = sp500_df[0]
```

```python
sp_df
```

```python
sp_ticker = list(sp_df['Symbol'].values)
```

```python
sp_ticker
```

```python
df = yf.download(sp_ticker)
```

```python
df
```

```python
df['Adj Close'][['A', 'AAL', 'AAP', 'AAPL', 'ABBV', 'ABC']].plot(subplots=True, figsize=[12,12])
```
